
# Formatted Input


A limited support for mathematical expressions (like in the wolfram's graphic interface) is supported. To insert a unicode character, press [Ctrl] + [Esc], then the (wolfram's) code and then [Ctrl] + [Esc] again to insert them.


The proper representation of certain codes like [Ctrl] + [Esc]dd[Ctrl] + [Esc], as well as
bidimensional expressions (exponents and indices)is still work in progress.

```mathematica
h[x]x 


Integrate[q[x],{q,-,}]

"2"
```

# Graphics

All the graphics commands are supported.

The format and the size of the graphics can be setted by the internal commands SetImageOutputFormat and 
SetImageWidth

```mathematica

Plot[x,{x,0,2}]


Plot3D[x y,{x,0,2},{y,0,2}]


GraphicsArray[{{Plot[x,{x,0,2}],Plot3D[x y,{x,0,2},{y,0,2}]}}]


ContourPlot[x y,{x,0,2},{y,0,2}]


DensityPlot[x y,{x,0,2},{y,0,2}]
```

```mathematica
Jupyter`SetImageOutputFormat["gif"];
Jupyter`SetImageWidth[200];

Plot[x,{x,0,2}]

Jupyter`SetImageOutputFormat["jpg"];
Jupyter`SetImageWidth[400];

Plot[x,{x,0,2}]


Jupyter`SetImageOutputFormat["svg"];
Jupyter`SetImageWidth[300];


Plot[x,{x,0,2},Prolog->{Text["g x",{1.0,1.0}]}]
```
